```python

#Scatter
#runs scored by 10 new players[100,50,91,78,89,25,34,19,9,10] wickets taken by same 10 new players [1,0,2,0,3,7,8,9,7,5]forms clusters for batsman and bowlers
runs=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(runs,wickets,color="violet",marker="d")#r,v,b,k,m,c,y....
plt.show
```

```python
#show the scores of two students in 5 subjects,subjects as x-axis and marks as y-axis
s1=[87,78,90,88,76]
s2=[60,70,67,99,73]
sub=["Maths","Science","Social","Telugu","English"]
plt.plot(sub,s1,label="student1")
plt.plot(sub,s2,label="student2")
plt.title("Subjects vs Marks")
plt.legend()
plt.show()
```

```python
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
sub=["Maths","Science","Social","Telugu","English"]
plt.subplot(2,1,1)
plt.plot(sub,s1,label="student1",color="k")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,s2,label="student2",color="b")
plt.legend()

```

```python
#Bar
A=[230,560,780,127,128]
B=[200,160,270,127,400]
print("Revenues of Company A",A)
print("Revenues of Company B",B)
profA=np.diff(A)
print("profits of A",profA)
profB=np.diff(B)
print("profits of B",profB)
year=["2019-2020","2020-2021","2021-2022","2022-2023"]
plt.subplot(2,1,1)
plt.bar(year,profA,color="k",label="profit A")
plt.legend()
plt.subplot(2,1,2)
plt.bar(year,profB,color="g",label="profit B")
plt.legend()
```

```python
Pie
a=np.array([25,60,5,10])
labe=["AIML","Python","Pandas","Numpy"]
plt.pie(a,labels=labe)
plt.show()
```

```python
#explode
a=np.array([25,60,5,10])
labe=["AIML","Python","Pandas","Numpy"]
explo=[0.2,0,0,0]
plt.pie(a,labels=labe,explode=explo,startangle=180,shadow="True")
plt.legend
plt.show()
```
